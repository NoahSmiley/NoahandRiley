```python
import random
import time
from IPython.display import clear_output
'''
A text based game of the Blackjack card game (21).
Developed by Noah Smiley and Riley Edmunds.
'''
class Deck():

    def __init__(self):
        self.deck = []
        self.suits = ['Hearts', 'Diamonds', 'Spades', 'Clubs']
        self.faces = [['Two',2],['Three',3],['Four',4],
                     ['Five',5],['Six',6],['Seven',7],['Eight',8],
                     ['Nine',9],['Ten',10],['Jack',10],['Queen',10],
                     ['King',10],['Ace',11]]

        #On instantiaion of the class, a deck of 52 cards is created.
        for suit in self.suits:
            for face in self.faces:
                x = (suit,face)
                self.deck.append(x)

    #Shuffle method to shuffle the deck.
    def shuffle(self):
        random.shuffle(self.deck)

    def select_card(self):
        return self.deck.pop()

    #Dunder str method to print the Deck out.
    def __str__(self):
        strings=[]
        for i in range(len(self.deck)):
            strings.append(f" The {self.deck[i][1][0]} of {self.deck[i][0]}")
        return '\n'.join(strings)

class GameLogic(Deck):

    def __init__(self):
        self.newDeck = Deck()
        self.newDeck.shuffle()
        self.deck= self.newDeck.deck
        self.dealerTotal = 0
        self.printCounter = 0

    def printSys(self,userCard,dealerCard,total):
        if self.printCounter == 0:
            print("Welcome to Noah and Riley's Blackjack game!\n")
            time.sleep(3)
            clear_output()  

        print("You were dealt: ", end ="                      ")
        print((f"The dealer was dealt:"))
        print(f"The {userCard[0][1][0]} of {userCard[0][0]}", end ="                     ")
        print((f"The {dealerCard[0][1][0]} of {dealerCard[0][0]}"))
        print((f"And The {userCard[1][1][0]} of {userCard[1][0]}"))

        if self.printCounter >=1 and len(userCard)>2:
            print(f"\nYou were dealt the {userCard[(len(userCard)-1)][1][0]} of {userCard[(len(userCard)-1)][0]}")

        print((f"Your current total is: {total}"))
        self.printCounter += 1

    def hit(self,userCard):
        result = input("Would you like to hit? y/n")
        if result == 'y' or result == 'Y':
            userCard.append(self.newDeck.select_card())
        else:
            pass
        return userCard

    def findTotal(self,userCard,):
        self.userTotal=0
        for i in range(len(userCard)):
            self.userTotal += userCard[i][1][1]
        return self.userTotal

    def new_round(self):
        self.userTotal = 0
        #Card hands for the dealer and user
        userCard = [self.newDeck.select_card(), self.newDeck.select_card()]
        dealerCard = [self.newDeck.select_card(), self.newDeck.select_card()]

        self.printSys(userCard,dealerCard,self.findTotal(userCard))
        time.sleep(3)
        self.hit(userCard)
        clear_output()
        self.printSys(userCard,dealerCard,self.findTotal(userCard))

def main():
    newGame = GameLogic()
    newGame.new_round()
main()
```
